```python
import pandas as pd
```

```python
accidentDF = pd.read_csv("../US_Accidents_Dec20_updated.csv")
```

```python
for col in accidentDF.columns:
    if accidentDF[col].isnull().sum() >= 100000:
        del accidentDF[col]
```

```python
accidentDF.isnull().sum()
```

```python
redundant_attributes = ["Weather_Timestamp", "Wind_Direction", "Timezone", "Airport_Code"]
for val in redundant_attributes:
    del accidentDF[val]
```

```python
accidentDF.isnull().sum()
```

```python
#these attributes are too important to drop outright, but have too many null values to just drop rows where
pitaValues = ["Temperature(F)", "Humidity(%)", "Pressure(in)", "Visibility(mi)", "Pressure(in)", "Weather_Condition"]
#these attributes have few enough null values that we can just drop the null values
weedableValues = ["City", "Zipcode", "Sunrise_Sunset", "Civil_Twilight", "Nautical_Twilight", "Astronomical_Twilight"]
```

```python
accidentDF["Temperature(F)"].fillna(method="ffill", inplace=True)
accidentDF["Humidity(%)"].fillna(method="ffill", inplace=True)
accidentDF["Visibility(mi)"].fillna(method="ffill", inplace=True)
accidentDF["Pressure(in)"].fillna(method="ffill", inplace=True)
accidentDF["Weather_Condition"].fillna(method="ffill", inplace=True)

accidentDF.isnull().sum()
```

```python
accidentDF = accidentDF.dropna()

accidentDF.isnull().sum()
```

```python
for col in accidentDF.columns:
    print(accidentDF[col][1])
```

```python
continuous = ['Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng', 'Distance(mi)', 'Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)']

#values that will not require further preprocessing to be analyzed
discrete_encodable = ['Side', 'City', 'County', 'State', 'Weather_Condition',
       'Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit',
       'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming',
       'Traffic_Signal', 'Turning_Loop', 'Sunrise_Sunset', 'Civil_Twilight',
       'Nautical_Twilight', 'Astronomical_Twilight']

#values that will require further preprocessing to be analyzed
discrete_unencodable = ['Start_Time', 'End_Time', 'Description', 'Street']
```

```python
accidentDF_scaled = accidentDF.copy()
```

```python
for col in continuous:
    accidentDF_scaled[col] = (accidentDF_scaled[col] - accidentDF_scaled[col].mean())/(accidentDF_scaled[col].std())

accidentDF_scaled[continuous]
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

for col in discrete_encodable:
    accidentDF_scaled[col] = label_encoder.fit_transform(accidentDF_scaled[col])
```
